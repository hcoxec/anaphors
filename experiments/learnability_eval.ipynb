```python
import torch
from discourse_lang import DataHandler
import json
import matplotlib.pyplot as plt
```

Plotting training and validation curves (with whatever checkpoints saved from running code in learnability_expts.ipynb)

```python
with open("learnability_final_checkpoints/receiver_null_checkpoints", 'r') as inf:
    dicts1 = json.load(inf)
```

```python
with open("learnability_final_checkpoints/receiver_comp_checkpoints", 'r') as inf:
    dicts2 = json.load(inf)
```

```python
with open("learnability_final_checkpoints/receiver_tok_checkpoints", 'r') as inf:
    dicts3 = json.load(inf)
```

```python
train_checkpoints_null = {}
val_checkpoints_null = {}
for n, i in enumerate(dicts1):
    for k in i.keys():
        training = i[k][0]
        val = i[k][1]
        for x in training:
            try:
                train_checkpoints_null[x[0]] += x[1]
            except KeyError:
                train_checkpoints_null[x[0]] = x[1]
        for x in val:
            try:
                val_checkpoints_null[x[0]] += x[1]
            except KeyError:
                val_checkpoints_null[x[0]] = x[1]
```

```python
train_checkpoints_null = {k:v/10 for k, v in train_checkpoints_null.items()}
print(train_checkpoints_null)
```

```python
val_checkpoints_null = {k:v/10 for k, v in val_checkpoints_null.items()}
print(val_checkpoints_null)
```

```python
train_checkpoints_comp = {}
val_checkpoints_comp = {}
for n, i in enumerate(dicts2):
    for k in i.keys():
        training = i[k][0]
        val = i[k][1]
        for x in training:
            try:
                train_checkpoints_comp[x[0]] += x[1]
            except KeyError:
                train_checkpoints_comp[x[0]] = x[1]
        for x in val:
            try:
                val_checkpoints_comp[x[0]] += x[1]
            except KeyError:
                val_checkpoints_comp[x[0]] = x[1]    
```

```python
train_checkpoints_comp = {k:v/10 for k, v in train_checkpoints_comp.items()}
print(train_checkpoints_comp)
```

```python
val_checkpoints_comp = {k:v/10 for k, v in val_checkpoints_comp.items()}
print(val_checkpoints_comp)
```

```python
train_checkpoints_tok = {}
val_checkpoints_tok = {}
```

```python
for n, i in enumerate(dicts3):
    for k in i.keys():
        training = i[k][0]
        val = i[k][1]
        for x in training:
            try:
                train_checkpoints_tok[x[0]] += x[1]
            except KeyError:
                train_checkpoints_tok[x[0]] = x[1]
        for x in val:
            try:
                val_checkpoints_tok[x[0]] += x[1]
            except KeyError:
                val_checkpoints_tok[x[0]] = x[1]    
```

```python
train_checkpoints_tok = {k:v/10 for k, v in train_checkpoints_tok.items()}
print(train_checkpoints_tok)
```

```python
val_checkpoints_tok = {k:v/10 for k, v in val_checkpoints_tok.items()}
print(val_checkpoints_tok)
```

```python
plt.rcParams["figure.figsize"] = (30,20)
```

```python
for d, n in zip((train_checkpoints_comp, train_checkpoints_tok, train_checkpoints_null), ("No Elision", "Pronoun", "Pro-drop")):# ("No Elision", "Pronoun", "Pro-drop")):
    lists = sorted(d.items(), key=lambda item: int(item[0]))[:30]
    new_list = []
    for num,item in enumerate(lists, 1):
        new_list.append((num, item[1]))
    x, y = zip(*new_list) # unpack a list of pairs into two tuples
    plt.plot(x, y, label = n, linewidth=3)
    plt.xlabel("Epoch", fontsize=40)
    plt.ylabel("Loss", fontsize=40)
    plt.xticks(range(0,31,5), fontsize=40)
    plt.yticks(fontsize=40)
    plt.legend(prop={'size': 40})
    #plt.savefig("ReceiverLangsTrainLossFinal_V2_largefont.pdf")
plt.show()
```
